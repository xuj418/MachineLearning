```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X = np.linspace(-4, 4, 100)
y = 0.5 * X ** 2 + X + 2 + np.random.normal(0, 1, size=100)
```

```python
plt.scatter(X, y)
```

```python
X = X.reshape(-1, 1)
```

### 使用线性回归预测

```python
linear = LinearRegression()
linear.fit(X, y)
```

```python
linear.score(X, y)
```

```python
plt.scatter(X, y)
plt.plot(X, linear.predict(X), color="r")
```

### 添加一个项

```python
X2 = np.hstack((X**2, X)) 
```

```python
X2.shape
```

```python
linear2 = LinearRegression()
linear2.fit(X2, y)
```

```python
linear2.score(X2, y)
```

```python
plt.scatter(X, y)
plt.plot(X, linear2.predict(X2), color="r")
```

```python
#对应方程的系数
linear2.coef_
```

```python
#对应方程的截距
linear2.intercept_
```
